```python
import pandas as pd
import numpy as np
```

```python
# Se cargan los csv
dfamazon = pd.read_csv('amazon_prime_titles-score.csv')
dfdisney = pd.read_csv('disney_plus_titles-score.csv')
dfhulu = pd.read_csv('hulu_titles-score.csv')
dfnetflix = pd.read_csv('netflix_titles-score.csv')
```

### Transformación 1

```python
# amazon

dfamazon['show_id'] = 'a' + dfamazon['show_id'].astype(str)
```

```python
dfamazon.head(3)
```

```python
# Transformación 1
# disney plus

dfdisney['show_id'] = 'd' + dfdisney['show_id'].astype(str)

```

```python
dfdisney.head(2)
```

```python
#  Transformación 1
# hulu

dfhulu['show_id'] = 'h' + dfhulu['show_id'].astype(str)
```

```python
dfhulu.head(1)
```

```python
# Trandformación 1
# Netflix

dfnetflix['show_id'] = 'n' + dfnetflix['show_id'].astype(str)
```

```python
dfnetflix.head(1)
```

### Transformación 2

```python
# Transformación 2, llenar los nulos del campo rating con una "G"

def Poner_G(dframeG):
    dframeG['rating'] = dframeG['rating'].fillna('G')
    
    return dframeG
```

```python
# función para todos

Poner_G(dfamazon)
Poner_G(dfdisney)
Poner_G(dfhulu)
Poner_G(dfnetflix)
```

## Transformación 3

```python
# Método

def Fomratear_fecha(dfentrada):
    dfentrada['date_added'] = pd.to_datetime(dfentrada.date_added)
    return dfentrada
```

```python
# Aplicando el método
Fomratear_fecha(dfamazon)
Fomratear_fecha(dfdisney)
Fomratear_fecha(dfhulu)
Fomratear_fecha(dfnetflix)
```

### Transformación 4

```python
def Convertir_a_minuscula(dfminuscula):
    dfminuscula = dfminuscula.apply(lambda col: col.str.lower() if col.dtype == "object" else col)
    return dfminuscula
```

```python
# Aplicar minusculas a todos los data frames

Convertir_a_minuscula(dfamazon)
Convertir_a_minuscula(dfdisney)
Convertir_a_minuscula(dfhulu)
Convertir_a_minuscula(dfnetflix)
```

### Transformación 5

```python
dfamazon.dtypes
```

```python
# dividir columnas

#df[['Dialling Code','Cell-Number']]=df.Number.str.split('-',expand=True)

def Dos_columnas(df_adividir):
    df_adividir[['duration_int', 'duration_type']] = df_adividir.duration.str.split(' ',expand=True)
    return df_adividir
```

```python
Dos_columnas(dfamazon)
Dos_columnas(dfdisney)
Dos_columnas(dfhulu)
Dos_columnas(dfnetflix)
```

#### Guardar cambios T4 

```python
amazon = Convertir_a_minuscula(dfamazon)
disney = Convertir_a_minuscula(dfdisney)
hulu = Convertir_a_minuscula(dfhulu)
netflix = Convertir_a_minuscula(dfnetflix)
```

### Transformación 5, tipo de datos en columnas duration_int y duration_type

```python
amazon.dtypes
```

```python
# Se crea la función:

def Convertir_dt(dframe):
    dframe["duration_int"] = dframe['duration_int'].astype(float).astype('Int64')
    dframe["duration_type"] = dframe["duration_type"].astype('str')
    
    return dframe
```

```python
Convertir_dt(amazon)
Convertir_dt(disney)
Convertir_dt(hulu)
Convertir_dt(netflix)



```

```python
netflix.dtypes
```

#### Algo de limpieza

```python
# quitar la 'b'  de duration type de amazon ¿?
amazon['duration_type'] = amazon['duration_type'].replace({'b':''}, regex=True)

# Cambiar 'seasons' por 'season' en todos los dataframes

def Normalizar_seasons(dataframe):
    dataframe['duration_type'] = dataframe['duration_type'].replace({'seasons':'season'}, regex=True)
    return dataframe   
```

```python
# Aplicamos en todos los df
Normalizar_seasons(amazon)
Normalizar_seasons(disney)
Normalizar_seasons(hulu)
Normalizar_seasons(netflix)
```
